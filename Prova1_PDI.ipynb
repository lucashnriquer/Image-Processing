```python
#Q1 - Conversao RGB-YIQ-RGB

import numpy as np
from PIL import Image

CaminhodoArquivo = '/content/Woman.png'

class Imagem:
    img = Image.open(CaminhodoArquivo).convert('RGB')                            #Abre a imagem p/ leitura


def RGBtoYIQtoRGB():
    imagemYIQ = np.array(Imagem.img.copy())  #Convertendo para o formato numpyArray
    imagemYIQ = imagemYIQ.astype(float)
    largura, altura = Imagem.img.size                           #Capturando altura e largura 

    for i in range(altura): 
        for j in range(largura):
            R,G,B = Imagem.img.getpixel((j, i))

            imagemYIQ[i][j][0] = 0.299*R + 0.587*G + 0.114*B
            imagemYIQ[i][j][1] = 0.596*R - 0.274*G - 0.322*B
            imagemYIQ[i][j][2] = 0.211*R - 0.523*G + 0.312*B

    imagemRGB = imagemYIQ.astype(int)

    for i in range(altura):
        for j in range(largura):                                    #Conversao para o formato de volta RGB                            
            R = 1.000*imagemYIQ[i][j][0] + 0.956 *imagemYIQ[i][j][1] + 0.621*imagemYIQ[i][j][2]      
            G = 1.000*imagemYIQ[i][j][0] - 0.272 *imagemYIQ[i][j][1] - 0.647*imagemYIQ[i][j][2]
            B = 1.000*imagemYIQ[i][j][0] - 1.106 *imagemYIQ[i][j][1] + 1.703*imagemYIQ[i][j][2]
            
            if R > 255:                                   #Verificando os limites de R, G e B na volta
                R = 255
            if R < 0:
                R = 0

            if G > 255:
                G = 255
            if G < 0:
                G = 0

            if B > 255:
                B = 255
            if B < 0:
                B = 0

            imagemRGB[i][j][0] = int(R);
            imagemRGB[i][j][1] = int(G);
            imagemRGB[i][j][2] = int(B);
    
    imagemRGB = np.uint8(imagemRGB)

    img = Image.fromarray(imagemRGB, 'RGB')
    img.save('/content/rgb.png')  
  
RGBtoYIQtoRGB()

```

```python
#Q2 - Negativo

from PIL import Image
import numpy as np

caminhoDoArquivo = "/content/Woman.png"

class Imagem:
    img = Image.open(caminhoDoArquivo).convert('RGB')                                     #Abre a imagem p/ leitura
    largura, altura = img.size                                  #Tupla com a altura e largura da imagem e seta os valores nas variaveis
     
def negative():
    out = Image.new('RGB', Imagem.img.size)                     #Cria o arquivo que vai salvar a imagem convertida

    for x in range(Imagem.largura):                             #Processo de inversao de cores da imagem
        for y in range(Imagem.altura):                
                R,G,B = Imagem.img.getpixel((x,y)) 
                Rpixel  = 255 - R
                Gpixel  = 255 - G
                Bpixel  = 255 - B      
                out.putpixel((x,y),(Rpixel, Gpixel, Bpixel)) 

    out.save('/content/negative.png')                           #Salvando a imagem em negativo

def negativeY():

    imagemYIQ = np.array(Imagem.img.copy())  #Convertendo para o formato numpyArray
    imagemYIQ = imagemYIQ.astype(float)

    for i in range(Imagem.altura): 
        for j in range(Imagem.largura):
            R,G,B = Imagem.img.getpixel((j, i))

            imagemYIQ[i][j][0] = 255 - (0.299*R + 0.587*G + 0.114*B)
            imagemYIQ[i][j][1] = 0.596*R - 0.274*G - 0.322*B
            imagemYIQ[i][j][2] = 0.211*R - 0.523*G + 0.312*B

    imagemRGB = imagemYIQ.astype(int)

    for i in range(Imagem.altura):
        for j in range(Imagem.largura):                                    #Conversao para o formato de volta RGB                            
            R = 1.000*imagemYIQ[i][j][0] + 0.956 *imagemYIQ[i][j][1] + 0.621*imagemYIQ[i][j][2]      
            G = 1.000*imagemYIQ[i][j][0] - 0.272 *imagemYIQ[i][j][1] - 0.647*imagemYIQ[i][j][2]
            B = 1.000*imagemYIQ[i][j][0] - 1.106 *imagemYIQ[i][j][1] + 1.703*imagemYIQ[i][j][2]
            
            if R > 255:                                   #Verificando os limites de R, G e B na volta
                R = 255
            if R < 0:
                R = 0

            if G > 255:
                G = 255
            if G < 0:
                G = 0

            if B > 255:
                B = 255
            if B < 0:
                B = 0

            imagemRGB[i][j][0] = int(R);
            imagemRGB[i][j][1] = int(G);
            imagemRGB[i][j][2] = int(B);
    
    imagemRGB = np.uint8(imagemRGB)

    img = Image.fromarray(imagemRGB, 'RGB')
    img.save('/content/negativeY.png')  

negative()
negativeY()
```

```python
#QUESTAO 3

import numpy as np
from PIL import Image

CaminhodoArquivo = "/content/Woman.png"

class Imagem:
    img = Image.open(CaminhodoArquivo).convert('RGB')                            #Abre a imagem p/ leitura                          

with open('variaveis.txt') as f:
    m = int(f.readline())
    n = int(f.readline())
    pivox = int(f.readline())
    pivoy = int(f.readline())
    offset = int(f.readline())

def FiltroMedia(filtro):
    imagem= np.array(Imagem.img.copy())  #Convertendo para o formato numpyArray
    imagem = imagem.astype(float)
    altura, largura, rgb = imagem.shape                           #Capturando altura e largura 


    if filtro == "media":
      mascara = np.zeros((m,n))
      media = 1/(m * n)

      for i in range(m):
          for j in range(n):
              mascara[i][j] = media

    if filtro == "sobelH":
      mascara = np.array([[-1,-2,-1],
                         [0, 0, 0],
                         [1, 2, 1]])
    if filtro == "sobelV":
      mascara = np.array([[1, 0,-1],
                         [2, 0, -2],
                         [1, 0, -1]])


    imagemSaida = np.zeros_like(imagem)

    for i in range(largura-m):
        for j in range(altura-n):

          valorPixelR=0
          valorPixelG=0
          valorPixelB=0

          for x in range(m):
              for y in range(n):
                  valorPixelR += mascara[y][x]*imagem[y+j][x+i][0]
                  valorPixelG += mascara[y][x]*imagem[y+j][x+i][1]
                  valorPixelB += mascara[y][x]*imagem[y+j][x+i][2]

          valorPixelR += offset
          valorPixelG += offset
          valorPixelB += offset

          if valorPixelR > 255:                                   #Verificando os limites de R, G e B
              valorPixelR = 255
          if valorPixelR < 0:
              valorPixelR = 0
          if valorPixelG > 255:
              valorPixelG = 255
          if valorPixelG < 0:
              valorPixelG = 0
          if valorPixelB > 255:
              valorPixelB = 255
          if valorPixelB < 0:
              valorPixelB = 0

          imagemSaida[j+pivoy][i+pivox][0] = valorPixelR 
          imagemSaida[j+pivoy][i+pivox][1] = valorPixelG
          imagemSaida[j+pivoy][i+pivox][2] = valorPixelB

    imagemSaida = np.uint8(imagemSaida)
    
    if filtro == "media":  
      img = Image.fromarray(imagemSaida, 'RGB')
      img.save('/content/media.png')  
    
    if filtro == "sobelV":

      maximoR = 0
      minimoR = 255

      maximoG = 0
      minimoG = 255

      maximoB = 0
      minimoB = 255

      for i in range(altura):
          for j in range(largura):

            if imagemSaida[i][j][0] < minimoR:
              minimoR = imagemSaida[i][j][0]

            if imagemSaida[i][j][0] > maximoR:
              maximoR = imagemSaida[i][j][0]

            if imagemSaida[i][j][1] < minimoG:
              minimoG = imagemSaida[i][j][1]

            if imagemSaida[i][j][1] > maximoG:
              maximoG = imagemSaida[i][j][1]
            
            if imagemSaida[i][j][2] < minimoB:
              minimoB = imagemSaida[i][j][2]

            if imagemSaida[i][j][2] > maximoB:
              maximoB = imagemSaida[i][j][2]

      for i in range(altura):
          for j in range(largura):

            imagemSaida[i][j][0] = round(((imagemSaida[i][j][0]-minimoR)*255)/(maximoR-minimoR))
            imagemSaida[i][j][1] = round(((imagemSaida[i][j][1]-minimoG)*255)/(maximoR-minimoG))
            imagemSaida[i][j][2] = round(((imagemSaida[i][j][2]-minimoB)*255)/(maximoR-minimoB))

      img = Image.fromarray(imagemSaida, 'RGB')
      img.save('/content/sobelV.png')  
            
    if filtro == "sobelH":

      maximoR = 0
      minimoR = 255

      maximoG = 0
      minimoG = 255

      maximoB = 0
      minimoB = 255

      for i in range(altura):
          for j in range(largura):

            if imagemSaida[i][j][0] < minimoR:
              minimoR = imagemSaida[i][j][0]

            if imagemSaida[i][j][0] > maximoR:
              maximoR = imagemSaida[i][j][0]

            if imagemSaida[i][j][1] < minimoG:
              minimoG = imagemSaida[i][j][1]

            if imagemSaida[i][j][1] > maximoG:
              maximoG = imagemSaida[i][j][1]
            
            if imagemSaida[i][j][2] < minimoB:
              minimoB = imagemSaida[i][j][2]

            if imagemSaida[i][j][2] > maximoB:
              maximoB = imagemSaida[i][j][2]

      for i in range(altura):
          for j in range(largura):

            imagemSaida[i][j][0] = round(((imagemSaida[i][j][0]-minimoR)*255)/(maximoR-minimoR))
            imagemSaida[i][j][1] = round(((imagemSaida[i][j][1]-minimoG)*255)/(maximoR-minimoG))
            imagemSaida[i][j][2] = round(((imagemSaida[i][j][2]-minimoB)*255)/(maximoR-minimoB))

      img = Image.fromarray(imagemSaida, 'RGB')
      img.save('/content/sobelH.png')

FiltroMedia("media")
```

```python
#QUESTAO 4

import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image

caminhoDoArquivo = "/content/Woman.png"

class Imagem:
     img = Image.open(caminhoDoArquivo).convert('RGB')                                     #Abre a imagem p/ leitura
     #largura, altura = img.size                                  #Tupla com a altura e largura da imagem e seta os valores nas variaveis

def mediana ():

    imagemYIQ = np.array(Imagem.img.copy())  #Convertendo para o formato numpyArray
    imagemYIQ = imagemYIQ.astype(float)
                                               #Tamanho da mascara
    m = 5
    n = 5
                                               #Tamanho da imagem
    A, L, rgb = imagemYIQ.shape 

    #----------------------------------------------------# 

    for i in range(A): 
        for j in range(L):
            R,G,B = Imagem.img.getpixel((j, i))

            imagemYIQ[i][j][0] = 0.299*R + 0.587*G + 0.114*B
            imagemYIQ[i][j][1] = 0.596*R - 0.274*G - 0.322*B
            imagemYIQ[i][j][2] = 0.211*R - 0.523*G + 0.312*B

    #----------------------------------------------------# 
                                              #Matriz pra armazenar a saida
    filtroMediana = np.zeros_like(imagemYIQ) 
                                               #Extensão com zeros
    exty = int((m) / 2)
    extx = int((n) / 2)
                                                         #Matriz extendida
    matrizSaida = np.zeros((A+2*exty, L+2*extx, rgb))

                                                         #Adicionando valores da imagem original na matrix extendida 
    matrizSaida[exty:A+exty, extx:L+extx] = imagemYIQ

                                                         #Redefinindo os pixels da imagem
    for y in range (exty , A+exty):
        for x in range (extx , L +extx):
            
            valoresY = matrizSaida[y-exty : y+exty + 1, x-extx : x+extx + 1, 0]
          
            medianaY = int(np.median(valoresY))                     #Aplicando o calculo da mediana com o numpy para valores Y          
             
            filtroMediana[y -exty,x -extx,0] = medianaY

    filtroMediana[:,:,1] = imagemYIQ[:,:,1]        
    filtroMediana[:,:,2] = imagemYIQ[:,:,2] 

    imagemRGB = filtroMediana.astype(int)

    for i in range(A):
        for j in range(L):                                    #Conversao para o formato de volta RGB                            
            R = 1.000*filtroMediana[i][j][0] + 0.956 *filtroMediana[i][j][1] + 0.621*filtroMediana[i][j][2]      
            G = 1.000*filtroMediana[i][j][0] - 0.272 *filtroMediana[i][j][1] - 0.647*filtroMediana[i][j][2]
            B = 1.000*filtroMediana[i][j][0] - 1.106 *filtroMediana[i][j][1] + 1.703*filtroMediana[i][j][2]
            
            if R > 255:                                   #Verificando os limites de R, G e B na volta
                R = 255
            if R < 0:
                R = 0

            if G > 255:
                G = 255
            if G < 0:
                G = 0

            if B > 255:
                B = 255
            if B < 0:
                B = 0

            imagemRGB[i][j][0] = int(R);
            imagemRGB[i][j][1] = int(G);
            imagemRGB[i][j][2] = int(B);
    
    imagemRGB = np.uint8(imagemRGB)
    
    plt.imsave('/content/filtromediana.jpg', imagemRGB)

mediana()
```

```python
#QUESTAO 5

import numpy as np 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy import signal
from sklearn.preprocessing import normalize

def buscaMascaraCorrelacao(flag):
    mascara = mpimg.imread("/content/Woman_eye.png")*255
    mascaraR = []
    mascaraG = []
    mascaraB = []

    for row in mascara:
        # mascara é uma lista de rows que possuem listas de pixels que são listas [R, G, B] cujos indices são [0, 1, 2]
        mascaraR.append([pixel[0] for pixel in row]) # Dá um append no R com uma lista que representa uma linha apenas com o valor de R
        mascaraG.append([pixel[1] for pixel in row]) # Dá um append no G com uma lista que representa uma linha apenas com o valor de G
        mascaraB.append([pixel[2] for pixel in row]) # Dá um append no B com uma lista que representa uma linha apenas com o valor de B

    mascaraNormalizadaR = normalize(np.array(mascaraR))
    mascaraNormalizadaG = normalize(np.array(mascaraG))
    mascaraNormalizadaB = normalize(np.array(mascaraB))

    #tamanho da mascara
    m, n, _ = mascara.shape

    imagem = mpimg.imread("/content/Woman.png")*255
    #pegando o tamanho da imagem
    A, L, rgb = imagem.shape
    
    mediaCorrelacoesRGB = np.zeros((A, L))

    #extensão com zeros
    exty = int((m) / 2)
    extx = int((n) / 2)

    #nova matriz extendida
    matrizExt = np.zeros((A+2*exty, L+2*extx, rgb))

    #adicionando valores da imagem original na matrix extendida 
    matrizExt[exty:A+exty, extx:L+extx] = imagem
    #print(matrizExt.shape)

    #pegando os pixels da imagem
    for y in range (exty , A+exty):
        for x in range (extx , L+extx):
            valoresR = normalize(matrizExt[y-exty : y+exty + 1, x-extx : x+extx + 1, 0])
            valoresG = normalize(matrizExt[y-exty : y+exty + 1, x-extx : x+extx + 1, 1])
            valoresB = normalize(matrizExt[y-exty : y+exty + 1, x-extx : x+extx + 1, 2])
            
            # mode='valid' define que o resultado da operação não operará em linhas e colunas extendidas por zero
            # será apenas a operação pura entre as duas matrizes
            # boundary='symm' significa que as bordas serão simétricas
            # É necessário pegar o elemento [0][0], pois o correlate2d apenas retorna uma lista de listas, mesmo
            # que o resultado seja apenas um elemento
            correlacaoR = signal.correlate2d(valoresR, mascaraNormalizadaR, mode='valid', boundary='symm')[0][0]
            correlacaoG = signal.correlate2d(valoresG, mascaraNormalizadaG, mode='valid', boundary='symm')[0][0]
            correlacaoB = signal.correlate2d(valoresB, mascaraNormalizadaB, mode='valid', boundary='symm')[0][0]

            mediaCorrelacoesRGB[y-exty,x-extx] = (correlacaoR + correlacaoG + correlacaoB) / 3

    # Encontra-se a maior correlação:

    maiorCorrelacao = 0
    maiorCorrelacaoX = 0
    maiorCorrelacaoY = 0

    for y in range(len(mediaCorrelacoesRGB)):
        for x in range(len(mediaCorrelacoesRGB[0])):
            if mediaCorrelacoesRGB[y][x] > maiorCorrelacao:
                maiorCorrelacao = mediaCorrelacoesRGB[y][x]
                maiorCorrelacaoX = x
                maiorCorrelacaoY = y

    if flag == 0:

      # Visualização da Imagem
      fig, ax = plt.subplots(1)
      ax.imshow(imagem/255) # Exibe a imagem no gráfico
      # É criado um retângulo para adicionar na imagem:
      rect = patches.Rectangle((maiorCorrelacaoX-extx, maiorCorrelacaoY-exty), n, m, linewidth=1,edgecolor='r',facecolor='none')
      ax.add_patch(rect) # Adiciona o retângulo na imagem
      plt.savefig("/content/corrmax.png") # Salva o que foi plotado


    if flag == 1:

      maiorCorrelacao2 = 0
      maiorCorrelacaoX2 = 0
      maiorCorrelacaoY2 = 0

      for y in range(len(mediaCorrelacoesRGB)):
          for x in range(len(mediaCorrelacoesRGB[0])):
              if (y>maiorCorrelacaoY-m and y<maiorCorrelacaoY+m) and (x>maiorCorrelacaoX-n and x<maiorCorrelacaoX+n):
                  continue
              if mediaCorrelacoesRGB[y][x] > maiorCorrelacao2:
                  maiorCorrelacao2 = mediaCorrelacoesRGB[y][x]
                  maiorCorrelacaoX2 = x
                  maiorCorrelacaoY2 = y

      # Visualização da Imagem
      fig, ax = plt.subplots(1)
      ax.imshow(imagem/255) # Exibe a imagem no gráfico
      # É criado um retângulo para adicionar na imagem:
      rect = patches.Rectangle((maiorCorrelacaoX2-extx, maiorCorrelacaoY2-exty), n, m, linewidth=1,edgecolor='r',facecolor='none')
      ax.add_patch(rect) # Adiciona o retângulo na imagem
      plt.savefig("/content/corr.png") # Salva o que foi plotado

buscaMascaraCorrelacao(1)
```
